# CNN1 = Full chromosome I

## Define and train the model

```python
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv1D, Conv2D, MaxPooling1D, AveragePooling1D
from keras.callbacks import ModelCheckpoint
from keras import backend as K
from keras.utils import np_utils, plot_model
import pickle
```

```python
SMALL_SIZE = 10
MEDIUM_SIZE = 15
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

**Import simulations** : genotypes and haplotypes, F = founders, d = descendants

```python
n = 16       # number of founders
p = 37525    # nb of SNPs
G = 20       # nb of generations
N = 50       # nb of recombining individuals in a generation after sampling
I = 50       # nb of iterations
```

```python
with open("Recombinations_haplotypes_n{0}_p{1}_G{2}_N{3}_I{4}.pkl".format(n,p,G,N,I),"rb") as handler:
    haplo = pickle.load(handler)
    F_haplo = np.array(haplo[:n])
    d_haplo = np.array(haplo[n:])
with open("Recombinations_genotypes_n{0}_p{1}_G{2}_N{3}_I{4}.pkl".format(n,p,G,N,I),"rb") as handler:
    geno = pickle.load(handler)
    F_geno = np.array(geno[:n])
    d_geno = np.array(geno[n:])

print(F_haplo.shape,d_haplo.shape,F_geno.shape,d_geno.shape)
```

```python
im, ax = plt.subplots(2,1,figsize=(16,4))
ax[0].imshow(F_geno, aspect = 100)
ax[0].set(xlabel = "SNPs", ylabel = "Founders", title = "Genotypes of the 16 founders")
ax[1].imshow(F_haplo, aspect = 100)
ax[1].set(xlabel = "SNPs", ylabel = "Founders", title = "Haplotypes of the 16 founders")
plt.show()
```

**Separate** 3/4 of data for training and 1/4 for testing.

```python
xtrain = d_geno[:3*I*N//4]
ytrain = d_haplo[:3*I*N//4]
xtest = d_geno[3*I*N//4:]
ytest = d_haplo[3*I*N//4:]
xtrain.shape, ytrain.shape, xtest.shape, ytest.shape
```

```python
fig, (ax1,ax2) = plt.subplots(2,1,figsize=(14,3))
ax1.imshow(xtrain)
ax2.imshow(ytrain)
fig.text(0.5, 0.06, 'SNPs', ha='center', va='center',fontsize=15)
fig.text(0.08, 0.5, 'Haplotype Genotype', ha='center', va='center', rotation='vertical',fontsize=15)
fig.text(0.5, 0.92, 'Expected prediction compared to input data for chromosomes from the training set', ha='center', va='center',fontsize=17)
plt.show()
```

**Input genotypes** : creating an "image" with the descendant x17 alterning with the 16 founders.

```python
def chromosome_alternate_inputs (array, F_geno, n):
    A = np.zeros((array.shape[0],array.shape[1],2*n+1))
    for k in range(len(array)) :
        individual = array[k]
        a = individual[:, np.newaxis]
        for l in range(len(F_geno)) :
            a=np.concatenate([a,F_geno[l:l+1].T,individual[:, np.newaxis]], axis = 1)
        A[k]=a
    return A
```

**Cut haplotypes** in 4 SNPs-long segments.

```python
def chromosome_segments_outputs (array):
    A = np.zeros((len(array),len(array[0])//4,16))
    for k in range(len(array)) :
        individual = array[k]
        A[k] = np.array([np_utils.to_categorical(segment[2]-1,16) for segment in np.reshape(individual[:len(individual)//4*4],(-1,4))])
    return A
```

```python
Xtrain = chromosome_alternate_inputs(xtrain, F_geno, n)
Ytrain = chromosome_segments_outputs(ytrain)
display(Xtrain.shape, Ytrain.shape)
Xtest = chromosome_alternate_inputs(xtest, F_geno, n)
Ytest = chromosome_segments_outputs(ytest)
display(Xtest.shape, Ytest.shape)
```

```python
plt.rc('xtick', labelsize=0)    # fontsize of the tick labels
plt.rc('ytick', labelsize=0)    # fontsize of the tick labels
plt.rc('axes', labelsize=10)    # fontsize of the x and y labels
im, ax = plt.subplots(33,1,figsize=(16,10))
ax[0].set(title = 'Construction of the input with 1 descendant x17 alterning with the 16 founders')
for k in range(len(ax)) :
    ax[k].imshow(np.expand_dims(Xtrain[0,:,k], axis = 1).T, aspect = 200)
    ax[k].set(ylabel = "{individual}".format(individual="D" if k//2 == k/2 else "F{}".format(k//2+1)))
plt.show()
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
```

```python
im, ax = plt.subplots(1,1,figsize=(16,6))
ax.imshow(Xtrain[0].T, aspect = 100)
ax.set(xlabel = "SNPs", ylabel = "Individuals", title = "Training genotypes for 1 descendant (x17) + 16 founders")
plt.show()
```

```python
im, ax = plt.subplots(1,1,figsize=(16,6))
ax.imshow(Ytrain[0].T,aspect=50, cmap = "Greys_r")
ax.set(xlabel = "SNPs window", ylabel = "Haplotypes", title = "Expected haplotypes prediction for a RIL")
plt.show()
```

**Define the model**

```python
K.clear_session()
model_1 = Sequential()
model_1.add(Conv1D(256, kernel_size=2,
                 activation='relu',
                 input_shape=(Xtest.shape[1], Xtest.shape[2]),
                 padding = 'same'))
model_1.add(Conv1D(128, kernel_size=2, activation='relu', padding = 'same'))
model_1.add(AveragePooling1D(pool_size=2,strides=2))
model_1.add(Dropout(0.25))
model_1.add(Conv1D(128, kernel_size=2, activation='relu', padding = 'same'))
model_1.add(AveragePooling1D(pool_size=2,strides=2))
model_1.add(Dropout(0.25))
model_1.add(Dense(128, activation='relu'))
model_1.add(Dropout(0.5))
model_1.add(Dense(128, activation='relu'))
model_1.add(Dropout(0.5))
model_1.add(Dense(16, activation='softmax'))
model_1.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])
filepath="CNN1_model_weights-improvement-{epoch:02d}-{val_acc:.2f}.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]
print(model_1.summary())
```

```python
plot_model(model_1, show_shapes=True, to_file='models/CNN1_model.png')
```

**Train the model** : not performed on the notebook

```python
epochs=100
model_1.fit(Xtrain, Ytrain,
          epochs=epochs,
          callbacks=callbacks_list,
          verbose=1,
          validation_data=(Xtest, Ytest))
```

**Separate** 3/4 of data for training and 1/4 for testing.

```python
xtrain = d_geno[:3*I*N//4]
ytrain = d_haplo[:3*I*N//4]
xtest = d_geno[3*I*N//4:]
ytest = d_haplo[3*I*N//4:]
xtrain.shape, ytrain.shape, xtest.shape, ytest.shape
```

**Input genotypes** : creating an "image" with the descendant x17 alterning with the 16 founders.

```python
def chromosome_alternate_inputs_bis (array, F_geno, linkage, n):
    A = np.zeros((array.shape[0],array.shape[1],3*n+1))
    for k in range(len(array)) :
        individual = array[k]
        a = individual[:, np.newaxis]
        for l in range(len(F_geno)) :
            a=np.concatenate([a,F_geno[l:l+1].T, linkage[:, np.newaxis], individual[:, np.newaxis]], axis = 1)
        A[k]=a
    return A
```

```python
def chromosome_segments_outputs (array):
    A = np.zeros((len(array),len(array[0])//4,16))
    for k in range(len(array)) :
        individual = array[k]
        A[k] = np.array([np_utils.to_categorical(segment[2]-1,16) for segment in np.reshape(individual[:len(individual)//4*4],(-1,4))])
    return A
```

**Input genotypes** : creating an "image" with the descendant x17 alterning with the 16 founders.

```python
def chromosome_alternate_inputs (array, F_geno, n):
    A = np.zeros((array.shape[0],array.shape[1],2*n+1))
    for k in range(len(array)) :
        individual = array[k]
        a = individual[:, np.newaxis]
        for l in range(len(F_geno)) :
            a=np.concatenate([a,F_geno[l:l+1].T,individual[:, np.newaxis]], axis = 1)
        A[k]=a
    return A
```

```python
K.clear_session()

first_input = Input(Xtest.shape[1:])
b1 = Conv1D(256, kernel_size=3, activation='relu', padding = 'same')(first_input)
b1 = Conv1D(128, kernel_size=3, activation='relu', padding = 'same')(b1)
b1 = AveragePooling1D(pool_size=2,strides=2)(b1)
b1 = Dropout(0.25)(b1)
b1 = Conv1D(128, kernel_size=3, activation='relu', padding = 'same')(b1)
b1 = AveragePooling1D(pool_size=2,strides=2)(b1)
b1 = Dropout(0.25)(b1)

second_input = Input(linkage_reduced.shape)
b2 = Dense(1, activation='relu', input_shape=linkage_reduced.shape)(second_input)
b2 = Dropout(0.25)(b2)

merged = Concatenate(axis = 2)([b1,b2])
output_layer = Dense(128, activation='relu')(merged)
output_layer = Dropout(0.5)(output_layer)
output_layer = Dense(128, activation='relu')(output_layer)
output_layer = Dropout(0.5)(output_layer)
output_layer = Dense(16, activation='softmax')(output_layer)

model_6 = Model(inputs=[first_input, second_input], outputs=output_layer)
model_6.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])
filepath="CNN6_model_weights-improvement-{epoch:02d}-{val_acc:.2f}.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]
print(model_6.summary())
```

```python
SMALL_SIZE = 10
MEDIUM_SIZE = 15
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
p = 1015    # p = nb of SNPs
n = 16      # n = nb of founders
G = 20      # nb of generations
N = 100     # number recombining individuals in a generation after sampling
I = 100     # nb of itrations
```

**Import simulations** : genotypes and haplotypes, F = founders, d = descendants

```python
with open("Recombinations_haplotypes_n{0}_p{1}_G{2}_N{3}_I{4}.pkl".format(n,p,G,N,I),"rb") as handler:
    haplo = pickle.load(handler)
    F_haplo = np.array(haplo[:n])
    d_haplo = np.array(haplo[n:])
with open("Recombinations_genotypes_n{0}_p{1}_G{2}_N{3}_I{4}.pkl".format(n,p,G,N,I),"rb") as handler:
    geno = pickle.load(handler)
    F_geno = np.array(geno[:n])
    d_geno = np.array(geno[n:])
with open("Linkage_p{}.pkl".format(p),"rb") as handler:
    linkage = pickle.load(handler)

print(F_haplo.shape, d_haplo.shape, F_geno.shape, d_geno.shape, linkage.shape)
```

**Separate** 3/4 of data for training and 1/4 for testing.

```python
xtrain = d_geno[:3*I*N//4]
ytrain = d_haplo[:3*I*N//4]
xtest = d_geno[3*I*N//4:]
ytest = d_haplo[3*I*N//4:]
xtrain.shape, ytrain.shape, xtest.shape, ytest.shape
```

**Input genotypes** : creating an "image" with the descendant x17 alterning with the 16 founders.

```python
def chromosome_alternate_inputs (array, F_geno, n):
    A = np.zeros((array.shape[0],array.shape[1],2*n+1))
    for k in range(len(array)) :
        individual = array[k]
        a = individual[:, np.newaxis]
        for l in range(len(F_geno)) :
            a=np.concatenate([a,F_geno[l:l+1].T,individual[:, np.newaxis]], axis = 1)
        A[k]=a
    return A
```

**Cut haplotypes** in 4 SNPs-long segments.

```python
def chromosome_segments_outputs (array):
    A = np.zeros((len(array),len(array[0])//4,16))
    for k in range(len(array)) :
        individual = array[k]
        A[k] = np.array([np_utils.to_categorical(segment[2]-1,16) for segment in np.reshape(individual[:len(individual)//4*4],(-1,4))])
    return A
```

```python
linkage_ext = np.expand_dims(linkage, axis = 1)
linkage_ext.shape
```

```python
K.clear_session()

first_input = Input(Xtest.shape[1:])
b1 = Conv1D(33, kernel_size=3, activation='relu', padding = 'same')(first_input)
b1 = Conv1D(33, kernel_size=3, activation='relu', padding = 'same')(b1)
b1 = AveragePooling1D(pool_size=2,strides=2)(b1)
b1 = Dropout(0.25)(b1)
b1 = Conv1D(33, kernel_size=3, activation='relu', padding = 'same')(b1)
b1 = AveragePooling1D(pool_size=2,strides=2)(b1)
b1 = Dropout(0.25)(b1)

second_input = Input(linkage_ext.shape)
b2 = Conv1D(33, kernel_size=3, activation='relu', padding = 'same')(second_input)
b2 = Conv1D(33, kernel_size=3, activation='relu', padding = 'same')(b2)
b2 = AveragePooling1D(pool_size=2,strides=2)(b2)
b2 = Dropout(0.25)(b2)
b2 = Conv1D(33, kernel_size=3, activation='relu', padding = 'same')(b2)
b2 = AveragePooling1D(pool_size=2,strides=2)(b2)
b2 = Dropout(0.25)(b2)
b2 = Dense(1, activation='relu')(b2)
b2 = Dropout(0.25)(b2)
b2 = Dense(1, activation='relu', input_shape=linkage_ext.shape)(b2)

merged = Concatenate(axis = 2)([b1,b2])
output_layer = Dense(33, activation='relu')(merged)
output_layer = Dropout(0.5)(output_layer)
output_layer = Dense(33, activation='relu')(output_layer)
output_layer = Dropout(0.5)(output_layer)
output_layer = Dense(16, activation='softmax')(output_layer)

model_7 = Model(inputs=[first_input, second_input], outputs=output_layer)
model_7.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])
filepath="CNN7_model_weights-improvement-{epoch:02d}-{val_acc:.2f}.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]
print(model_7.summary())
```
